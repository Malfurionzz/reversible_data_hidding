```python
import numpy as np
import pandas as pd
import cv2


import warnings; warnings.filterwarnings(action='once')

from skimage.metrics import mean_squared_error as mse
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim


import matplotlib.pyplot as plt

from genNeedImg import genNeedImg
from embedding import embedding
from extraction import extracting

%matplotlib inline
```

```python
# Import Data

cover_psnr = pd.read_csv("results/cover_psnr.csv",header=None)
cover_mse= pd.read_csv("results/cover_mse.csv",header=None)
cover_ssim = pd.read_csv("results/cover_ssim.csv",header=None)


embed_mse = pd.read_csv("results/embed_mse.csv",header=None)
embed_ssim = pd.read_csv("results/embed_ssim.csv",header=None)

recover_mse= pd.read_csv("results/recover_mse.csv",header=None)
recover_ssim = pd.read_csv("results/recover_ssim.csv",header=None)

spaceList=pd.read_csv("results/spaceList.csv",header=None)
import seaborn as sns
sns.set(color_codes=True)
```

```python
g=sns.kdeplot(spaceList[0], fill=True, color="r", label="Hidden Space")
g.set_xlabel('Hiden Space(%)')
g.set_title('Hiden Space Ratio', fontsize=22)
```

```python

# psnr.sort_values(axis=1,ascending=False)
# Draw Plot
fig, axes = plt.subplots(1, 3,figsize=(12, 3), constrained_layout=True)
g=sns.kdeplot(cover_psnr[0], fill=True, color="r", label="PSNR",ax=axes[0])

g.set_xlim(0,75)
g.set_xlabel('PSNR')
g.set_title('PSNR', fontsize=22)


# plt.figure(figsize=(8,5), dpi= 100)
g=sns.kdeplot(cover_mse[0], fill=True, color="r", label="MSE",ax=axes[1])
g.set_xlim(0,1.2)
g.set_xlabel('MSE')
g.set_title('MSE', fontsize=22)
# 核密度估计
# plt.figure(figsize=(8,5), dpi= 100)
g=sns.kdeplot(cover_ssim[0], fill=True, color="r", label="SSIM",ax=axes[2])
g.set_xlim(0.993,1)
g.set_xlabel('SSIM')
g.set_title('SSIM', fontsize=22)
plt.suptitle("KDE of metrics(cover,embed)",fontsize='30')
```

```python

# psnr.sort_values(axis=1,ascending=False)
# Draw Plot
embed=pd.concat([embed_ssim,embed_mse],axis=1,join="inner")
embed.columns =["SSIM", "MSE"] 
embed.value_counts()
```

```python
fig, axes = plt.subplots(1, 2,figsize=(12, 3), constrained_layout=True)

# plt.figure(figsize=(8,5), dpi= 100)
g=sns.scatterplot(x=embed_mse[0],y=embed_mse[0],data=embed_mse[0], color="r", label="mse",ax=axes[0])

g.set_xlabel('MSE')
g.set_title('MSE', fontsize=22)
g.set_xlim(-0.1,0.1)
g.set_ylim(-0.1,0.1)
g.set_xlabel('MSE')
g.set_ylabel('MSE')
# plt.figure(figsize=(8,5), dpi= 100)
g=sns.scatterplot(x=embed_ssim[0],y=embed_ssim[0],data=embed_ssim[0], color="r", label="SSIM",ax=axes[1])

g.set_xlabel('SSIM')
g.set_title('SSIM', fontsize=22)
g.set_xlabel('SSIM')
g.set_ylabel('SSIM')
g.set_xlim(0.9,1.1)
g.set_ylim(1.1,0.9)
plt.suptitle("Metrics(hiden,extracted)",fontsize=30)
```

```python
recover=pd.concat([recover_ssim,recover_mse],axis=1,join="inner")
recover.columns =["SSIM", "MSE"] 
recover.value_counts()
```

```python
fig, axes = plt.subplots(1, 2,figsize=(12, 3), constrained_layout=True)

# plt.figure(figsize=(8,5), dpi= 100)
g=sns.scatterplot(x=recover_mse[0],y=recover_mse[0],data=recover_mse[0], color="r", label="mse",ax=axes[0])

g.set_xlabel('MSE')
g.set_title('MSE', fontsize=22)
g.set_xlim(-0.01,0.01)
g.set_ylim(-0.01,0.01)
g.set_xlabel('MSE')
g.set_ylabel('MSE')
# plt.figure(figsize=(8,5), dpi= 100)
g=sns.scatterplot(x=recover_ssim[0],y=recover_ssim[0],data=recover_ssim[0], color="r", label="SSIM",ax=axes[1])

g.set_xlabel('SSIM')
g.set_title('SSIM', fontsize=22)
g.set_xlabel('SSIM')
g.set_ylabel('SSIM')
g.set_xlim(0.999999,1.000001)
g.set_ylim(0.999999,1.000001)
plt.suptitle("Metrics(cover,recover)",fontsize=30)
```
